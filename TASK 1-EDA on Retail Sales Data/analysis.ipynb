# Imputation of Missing Values

```python
df['Courier Status']=df['Courier Status'].fillna(df[df['Courier Status'].notnull()]['Courier Status'].mode()[0])
```

```python
df['currency']=df['currency'].fillna('INR')
```

```python
df['Amount']=df['Amount'].fillna(df['Amount'].mean())
```

```python
df['ship-city']=df['ship-city'].fillna(df[df['ship-city'].notna()]['ship-city'].mode()[0])
```

```python
df['ship-state']=df['ship-state'].fillna(df[df['ship-state'].notna()]['ship-state'].mode()[0])
```

```python
df['ship-state']=df['ship-state'].fillna(df[df['ship-state'].notna()]['ship-state'].mode()[0])
```

```python
df['ship-postal-code']=df['ship-postal-code'].fillna(df[df['ship-postal-code'].notna()]['ship-postal-code'].mode()[0])
```

```python
df['ship-country']=df['ship-country'].fillna("IN")
```

```python
df['promotion-ids']=df['promotion-ids'].fillna(df[df['promotion-ids'].notna()]['promotion-ids'].mode()[0])
```

```python
df = df.drop(columns=['Unnamed: 22','fulfilled-by'])
```

# Time Series Analysis: plot monthly and quarterly sales trends using line charts

```python
df['Day']=df["Date"].dt.day
df['Month']=df["Date"].dt.strftime('%b')
df['Year']=df["Date"].dt.year
```

```python
Monthly_sales = df.groupby('Month')['Amount'].sum()
Monthly_sales
```

```python
monthly_sales = Monthly_sales.reindex(['Mar', 'Apr', 'May', 'Jun'])

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line",marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)

plt.show()
```

Monthly Sales Trend Analysis:
The line chart shows that total sales increased sharply from March to April, with April recording the highest sales during the observed period. After April, sales declined gradually in May and June. Overall, the trend suggests that the business achieved its strongest performance in April, followed by a moderate decrease in sales over the next two months.

```python
df['Quarter'] = df['Date'].dt.to_period('Q')
```

```python
quarterly_sales = df.groupby('Quarter')['Amount'].sum()
quarterly_sales
```

```python
plt.figure(figsize=(10,5))

quarterly_sales.plot(kind='line', marker='o')

plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')

plt.grid(True)

plt.show()
```

Quarterly Sales Trend Analysis:
The quarterly sales trend shows that almost all sales were recorded during Q2 (April–June 2022), while Q1 (January–March 2022) had negligible sales. This pattern occurs because the dataset primarily covers transactions from April to June. Therefore, quarterly analysis provides limited insights, whereas monthly analysis is more appropriate for understanding sales trends in this dataset.

```python
top_products = (
    df.groupby('SKU')['Qty']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_products)
```

```python
plt.figure(figsize=(12,6))

top_products.plot(kind='bar')

plt.title('Top 10 Best-Selling Products')
plt.xlabel('SKU')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)

plt.grid(axis='y')

plt.show()
```

The analysis identifies the top 10 products based on the total quantity sold. These products contribute significantly to total sales volume and represent the company's highest-demand items. Focusing inventory and promotional strategies on these products can improve customer satisfaction and increase overall revenue.

```python
category_revenue = (
    df.groupby('Category')['Amount']
      .sum()
      .sort_values(ascending=False)
)

print(category_revenue)
```

```python
plt.figure(figsize=(10,6))

category_revenue.plot(kind='barh')

plt.title('Revenue by Product Category')
plt.xlabel('Revenue')
plt.ylabel('Category')

plt.show()
```

The bar chart shows the total revenue generated by each product category. Categories with the highest revenue are the major contributors to overall sales, while lower-performing categories may require promotional efforts or inventory optimization.

# Heatmap: correlation matrix between numerical variables

```python
numeric_df = df[['Qty', 'Amount']]
```

Correlation Analysis:

# At least one additional visualisation of your choice that reveals a non-obvious insight

```python
state_sales = (
    df.groupby('ship-state')['Amount']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(8,5))

state_sales.plot(kind='barh')

plt.title('Top 10 States by Revenue')
plt.xlabel('Revenue')
plt.ylabel('State')

plt.show()
```

Maharashtra, Karnataka, and Telangana contribute the highest revenue, suggesting these are the company's strongest markets.

```python
avg_order = df.groupby('Category')['Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))

avg_order.plot(kind='bar')

plt.title('Average Order Value by Category')
plt.xlabel('Category')
plt.ylabel('Average Revenue per Order')

plt.xticks(rotation=45)

plt.show()
```

The analysis reveals that Sets have the highest average order value (approximately ₹820), followed by Sarees and Western Dresses. This suggests that these categories consist of higher-priced products and contribute more revenue per customer order. On the other hand, Dupattas and Bottoms have the lowest average order values, indicating they are lower-priced products or are commonly purchased as individual items. This insight can help the business prioritize premium categories for marketing while using lower-value products in bundle offers or cross-selling strategies to increase the average order value.

# Conclusion & Business Recommendations

Conclusion :
The exploratory data analysis of the Amazon Sales Report provides valuable insights into sales performance, product categories, and customer purchasing behavior. Sales peaked in April and gradually declined in May and June, indicating a need to sustain demand after peak periods. Product categories such as Sets, Sarees, and Western Dresses generated the highest average order values, while categories like Dupatta and Bottom contributed lower revenue per order. The analysis also highlighted the best-selling products and revenue-generating categories, which can support data-driven business decisions.

Business Rcommendations :

1.Focus Marketing on High-Value Categories

 Finding: Sets, Sarees, and Western Dresses have the highest average order value.

 Recommendation: Allocate more advertising budget and promotional campaigns to these premium categories.
 Highlight these products during seasonal sales and festivals to maximize revenue.

 Expected Benefit: Higher revenue per customer order and improved overall profitability.

2.Increase Sales During Low-Performing Months

Finding: Sales declined after April.

Recommendation: Launch discounts, coupon campaigns, or festive promotions during May and June.
Use email marketing and personalized product recommendations to re-engage customers.

Expected Benefit: Reduced seasonal decline and more consistent monthly sales.

3.Bundle Low-Value Products with Premium Products

Finding: Dupattas and Bottoms have lower average order values.

Recommendation: Offer bundle deals such as:
Set + Dupatta, 
Kurta + Bottom, 
Saree + Accessories, 
Recommend complementary products during checkout.

Expected Benefit: Increased average order value and better sales of lower-priced items.

4.Maintain Inventory of Best-Selling Products

Finding: The top-selling SKUs contribute significantly to overall sales.

Recommendation: Ensure sufficient inventory for these products.
Use demand forecasting to avoid stock-outs during peak demand.

Expected Benefit: Higher customer satisfaction and fewer lost sales.

5.Prioritize High-Revenue Regions

Finding: A few states generate the highest revenue.

Recommendation: Focus marketing, logistics, and inventory planning in these high-performing regions while developing strategies to grow sales in lower-performing states.

Expected Benefit: Better resource allocation and increased regional sales.
